## Terminology

_**Defnition**_ A vector $\mathbf{v}$ is called a **probability vector** if
- its entries are nonnegative (they *could be zero))
- the sum of the entries are equal to 1.


- A square matrix $P$ is called a **stochastic matrix** if its colums are *probability vectors*

### Examples

- $\mathbf{v}= \begin{pmatrix} 0.3\\ 0.3 \\0.2 \\ 0.2 \end{pmatrix}$ is a probability vector.


- the matrix $P= \begin{pmatrix} 0.3 & 0.4 & 0.24\\ 0.3 & 0.6 & 0.16 \\0.1 & 0.0 & 0.6 \end{pmatrix}$ is a Stochastic matrix.

### Markov Chains

A **Markov Chain** is a sequence of probability vectors:

$\left\{ \mathbf{v_0}, \mathbf{v_1}, \ldots \mathbf{v_k}, \mathbf{v_{k+1} \ldots}  \right\}$

together with a matrix $M$, such that

$\mathbf{v_{k+1}}=M\mathbf{v_k}$

- The vectors $\mathbf{v_k}$ are called state vectors,
- The matrix is called the **migration matrix**, or **transition matrix**

*italicized text* ### Regular Stochastic Matrix
  A stochastic matrix $P$ is called **regular** if for some $k \geq 1$, all of the entries of $P^k$ are strictly positive.

 ### Steady-state vector
If $M$ is an $n \times n$ stochastic matrix, then a nonzero vector $\mathbf{v} \in R^n$ is called a **steady-state** vector or an **equilibrium** vector for $M$ if it has the following two properties :
- $\mathbf{v}$ is a probability vector, i.e., its entries are non-negative and the sum of entries of $\mathbf{v}$ equals 1.
- $M\mathbf{v} = \mathbf{v}$, i.e., $\mathbf{v}$ is an eigenvector of M corresponding to eigenvalue λ = 1

### Existence of Equilibrium (Steady-state)  vectors

_**Theorem**_

If $M$ is a regular stochastic matrix , then it has a *unique* stady-state vector $\mathbf{v}$.

Moreover, if

$\left\{ \mathbf{v_0}, \mathbf{v_1}, \ldots \mathbf{v_k}, \mathbf{v_{k+1} \ldots}  \right\}$

is a Markov Chain with initial state $\mathbf{v_0}$, and transition matrix $M$, then

$\displaystyle \lim_{n\to \infty }\mathbf{v_i} = \mathbf{v}$.

 This is equivallent to

$\displaystyle \lim_{n \to \infty} M^n ( \mathbf{v_0})= \mathbf{v}$


If the migration matrix, $M$,  of a Markov chain is regular , then
- the long term behavior of the Markov chain is decribed by the steady-state vector
- The eigenspace of $M$ corresponsding to the eigenvalue $\lambda = 1$ is one-dimensional
- The steady-state  vector is the unique probability vector in the eigenspace $E_1$, corresponding to the eigenvalue $\lambda=1$
- the long-term behavior does not depend on the initial state.

## Problem 3: Website search
---

Markov chains play an important role in online search.

“PageRank is an algorithm used by Google Search to rank websites in their search engine results. PageRank was named after Larry Page, one of the founders of Google. PageRank is a way of measuring the importance of website pages”
Source: https://en.Wikipedia.org/wiki/PageRank under the Creative Commons Attribution-ShareAlike License;

The theory behind PageRank is that pages that are linked to more often are more important and useful; identifying those that are linked to more often about a topic helps identify the pages that should be presented as most pertinent in a search.

In real world search, there are thousands or millions of pages linking together, resulting in huge transition matrices. Because of the size and other properties of these matrices, the mathematics behind PageRank is more sophisticated than the small example we examine here with only four websites. However our example is adequate to convey the main concept of PageRank and its use in search algorithms.

It should be noted that real world search algorithms, PageRank or similar Markov chain ranking schemes are only one of a variety of methodsused.
